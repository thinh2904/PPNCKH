```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data.csv',index_col='Year')
df
```

```python
plt.figure(figsize=(10,8))
plt.plot(df['Khá giỏi'])
plt.plot(df['Học lại'])
plt.plot(df['Học bổng'])
plt.title('TỈ LỆ KHÁ GIỎI, HỌC LẠI, HỌC BỔNG QUA 3 NĂM')
plt.ylabel('Tỉ lệ (%)')
plt.xlabel('Năm')
plt.legend(['Khá giỏi','Học lại','Học bổng'])
plt.show()
```

```python
plt.figure(figsize=(10,8))
N = 3
ind = np.arange(N) 
width = 0.25
  
xvals = df['Khá giỏi'].tolist()
bar1 = plt.bar(ind, xvals, width, color = 'r')
  
yvals = df['Học lại'].tolist()
bar2 = plt.bar(ind+width, yvals, width, color='g')
  
zvals = df['Học bổng'].tolist()
bar3 = plt.bar(ind+width*2, zvals, width, color = 'b')
  
plt.xlabel("Năm học")
plt.ylabel('Tỉ lệ (%)')
plt.title("TỈ LỆ KHÁ GIỎI, HỌC LẠI, HỌC BỔNG QUA 3 NĂM")
  
plt.xticks(ind+width,['2016-2017', '2017-2018', '2018-2019'])
plt.legend( (bar1, bar2, bar3), ('Khá giỏi', 'Học lại', 'Học bổng') )
plt.show()
```
